```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y', color='purple', alpha=0.5, linestyle='--', linewidth=5)   # x 방향 격자 무늬 추가
```
