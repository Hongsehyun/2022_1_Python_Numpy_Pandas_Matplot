# Matplotlib

 다양한 형태의 그래프를 통하여 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```


# 01. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y)
```

```python
# Plotting할 때 생기는 위 글자가 안나오게 하고 싶을 경우
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('그래프')   # 별다른 설정 없이 한국어를 입력하면 에러가 발생한다.
```

## 한글 폰트 설정

```python
# 한글 폰트 사용 설정 방법
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결


#matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac에서 한글 사용
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist   # 사용 가능한 폰트 확인
```

```python
[f.name for f in fm.fontManager.ttflist]   # 사용가능한 폰트의 이름만 가져오기
```

```python
plt.plot(x,y)
plt.title('그래프')
```

```python
# 별다른 설정 없이 -와 같은 특수문자를 입력하면 에러가 발생한다.
plt.plot([-1,0,1], [-5,-1,2])
```

```python
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
matplotlib.rcParams['axes.unicode_minus'] = False

plt.plot([-1,0,1], [-5,-1,2])
```
