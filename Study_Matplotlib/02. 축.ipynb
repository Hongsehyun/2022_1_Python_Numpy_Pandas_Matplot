# 02. 축



```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y)

# 개별 폰트 설정(fontdict) : Dictionary 형태로 적어주며
# family는 어떤 폰트를 쓸지 정의해주는 것, size는 글자의 크기를 정의해주는 것이다.
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20} )
plt.show()
```

```python
# x축과 y축 설정
plt.plot(x,y)
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='#00aa00')   # R G B 순서대로 00 00 00 인덱스 값 입력
```

```python
# x축과 y축 위치 설정
plt.plot(x,y)
plt.xlabel('x축', color='red', loc='right')    # left, center, right
plt.ylabel('y축', color='#00aa00', loc='top')  # top, center, bottom
```

```python
# x축과 y축 값 설정
plt.plot(x,y)
plt.xticks( [1,2,3] )
plt.yticks( [3,6,9,12] )
plt.xlabel('x축', color='red', loc='right')    # left, center, right
plt.ylabel('y축', color='#00aa00', loc='top')  # top, center, bottom
```
