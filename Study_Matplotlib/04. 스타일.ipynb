# 04. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y, label='무슨 데이터')
plt.legend()
```

```python
# 선 두껍게
plt.plot(x,y,linewidth=5)
```

## marker

```python
# 데이터 포인트 위치 표시[1]
plt.plot(x,y,marker='o')
```

```python
# 데이터 포인트 위치 표시[2]
plt.plot(x,y,marker='o', linestyle='None')
```

```python
# 데이터 포인트 위치 표시[3]
plt.plot(x,y,marker='X', markersize=15, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

## 색깔

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='#ff0000')
```

***

## 포맷

```python
plt.plot(x,y,'ro--')   # 색깔, 마커, 선 스타일을 한 번에 지정하는 방법
```

```python
plt.plot(x,y,'go')   # 색깔, 마커, 선 스타일을 한 번에 지정하는 방법
```

## 축약어

```python
plt.plot(x,y,marker='X', ms=15, mec='red', mfc='yellow')
```

## 투명도

```python
plt.plot(x,y,marker='X', ms=15, mec='red', mfc='yellow', alpha=0.5)   # 투명도 [0~1 사이의 값]
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
