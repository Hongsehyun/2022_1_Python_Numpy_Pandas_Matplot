# 06. 텍스트

텍스트는 그래프내에서 원하는 글자를 집어넣는 작업

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, marker='o')

for idx, txt in enumerate(y):      # enumerate를 쓰면 y좌표 데이터를 인덱스와 값으로 불러온다.
                                    # 즉, inx에는 0 1 2가,   txt에는 2 4 8이 순서대로 입력된다.
        plt.text(x[idx],y[idx], txt)
```

```python
plt.plot(x,y, marker='o')

for idx, txt in enumerate(y):
        plt.text(x[idx],y[idx]+0.3, txt, ha='center', color='b')
                # 숫자가 읽기 힘든 곳에 있으므로 y축 기준으로는 위로 좀 올리고
                # x축 기준으로는 가운데 정렬을 시키는 작업임!
```
