# 08. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors)
```

```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
# y축 범위를 수정하여 데이터 크기 비교를 명확하게 하기
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키

plt.bar(labels,values)
plt.ylim(175,200)
```

```python
# 막대 그래프를 날씬하게 출력해보기
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키

plt.bar(labels,values, width=0.4)
plt.ylim(175,200)
```

```python
# 글자끼리 서로 겹치지 않도록 하는 방법
plt.bar(labels,values, width=0.4)
plt.xticks(rotation=45)   # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45)   # y축의 이름 데이터 각도를 45도로 설정
```

```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키
ticks = ['1번학생','2번학생','3번학생']

plt.bar(labels,values)

plt.xticks(labels,ticks)   # x좌표 레이블 내용 변경하는 방법
plt.xticks(rotation=90)   # x축의 이름 데이터 각도를 45도로 설정
```
