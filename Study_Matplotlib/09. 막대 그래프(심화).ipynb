```python
labels = ['강백호', '서태웅', '정대만']   #이름
values = [190, 197, 194]   #키

plt.barh(labels, values)
plt.xlim(185,200)
```

```python
# 막대 안에 패턴 넣기
bar = plt.bar(labels, values)

bar[0].set_hatch('*')   # 첫번째 bar
bar[1].set_hatch('x')   # 두번째 bar
bar[2].set_hatch('..')   # 세번째 bar
```

```python
bar = plt.bar(labels, values)
plt.ylim(185,200)

for idx, rect in enumerate(bar):
        plt.text(idx,values[idx]+0.5, values[idx],ha='center', color='k')
```
