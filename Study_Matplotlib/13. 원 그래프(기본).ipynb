# 13. 원 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 원 그래프를 그리기 위해서는 Value와 lables에 대한 정보가 주어져야 한다.
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values) # values만 주어지면 pie chart가 그려지기는 하지만, 정보는 뜨지 않는다.
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=label)   # labels라는 속성에 label값을 넣어서 각 파이 영역별 어떤 정보인지 뜨게 한다.
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=label, autopct = '%.1f')   # auto percent = autopct   각 파이 영역별 몇 %를 차지하는지 표시해준다.
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=label, autopct = '%.1f%%')     # %까지 출력하고 싶다면 '%%'추가하기!
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90)     # 데이터가 차지하는 영역이 반시계방학으로 돌아가면서 생기는데
plt.show()                                                            # startangle이라는 명령어를 통해 시작 위치를 바꿀 수 있다.
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False)     
plt.show()   # 데이터가 차지하는 영역이 반시계방학으로 돌아가면서 생기는데 counterclock 명령어를 통해 방향을 바꿀 수 있다.
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

explode = [0.2, 0.1, 0, 0, 0, 0]                                         # 파이 데이터 영역을 띄울 때 사용
plt.pie(values, labels=label, autopct = '%.1f%%', explode=explode)

plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

explode = [0.1] * 6                                     
plt.pie(values, labels=label, autopct = '%.1f%%', explode=explode)
plt.legend()                                                                # 범례 위치가 맘에 들지 않음.

plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

explode = [0.1] * 6                                     
plt.pie(values, labels=label, autopct = '%.1f%%', explode=explode)
plt.legend( loc = (1.2, 0.3), title = '언어별 선호도' )           # 범례 위치가 수정 및 제목 추가.

plt.show()
```
