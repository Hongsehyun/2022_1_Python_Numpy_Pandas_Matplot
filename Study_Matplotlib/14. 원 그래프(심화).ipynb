# 14. 원 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
label = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

# 그래프 색상 부여
color = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.15] * 6

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False, colors=color, explode=explode) 
plt.show()
```

```python
# 원 그래프의 스타일 공부하기[1] - 원 그래프 형식 바꾸기
wedgeprops = {'width':0.8}     # 파이의 크기(width)를 0~1 사이의 값으로 지정하여 크기 설정 가능! 

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False, colors=color, explode=explode, wedgeprops=wedgeprops) 
plt.show()
```

```python
# 원 그래프의 스타일 공부하기[1] - 원 그래프 형식 바꾸기
wedgeprops = {'width':0.2}

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False, colors=color, explode=explode, wedgeprops=wedgeprops) 
plt.show()
```

```python
# 원 그래프의 스타일 공부하기[2] - explode로 인해 데이터가 부자연스럽게 보이는 문제 해결하기
wedgeprops = {'width':0.6, 'edgecolor':'w', 'linewidth':3}

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False, colors=color, wedgeprops=wedgeprops) 
plt.show()

# explode를 쓰면, 데이터끼리 서로 띄우는 과정에서 높이가 맞지 않게되고, 보기에 불편한 경우가 있을 수 있는데,
# 이 경우, explode대신에 edgecolor, linewidth 명령어를 이용함으로써 보다 그래프를 이쁘게 만들 수 있다.
```

```python
# 원 그래프의 스타일 공부하기[2] - explode로 인해 데이터가 부자연스럽게 보이는 문제 해결하기
wedgeprops = {'width':0.6, 'edgecolor':'k', 'linewidth':2}

plt.pie(values, labels=label, autopct = '%.1f%%', startangle=90, counterclock=False, colors=color, wedgeprops=wedgeprops) 
plt.show()
```

```python
# 원 그래프의 스타일 공부하기[3] - 글자 겹치는 문제 해결하기 - 함수를 만들어야 함! 10% 이상의 값만 출력되게 하면 된다.
def custom_autopct(pct):
    return ('%0.1f%%' % pct) if pct >= 10 else ''   # pct정보를 받았을 때, pct가 10 이상이면 값을 return하고
                                                      # 10 아래면 값을 ''로 즉, 빈칸으로 준다.

plt.pie(values, labels=label, autopct = custom_autopct, startangle=90, counterclock=False, colors=color, wedgeprops=wedgeprops) 
plt.show()
```

```python
# 원 그래프의 스타일 공부하기[4] - 퍼센트 글자 위치 조정하기
# 명령어 pctdistance :: 0과 1 사이의 숫자 입력, 0에 가까울 수록 중앙에 가깝다.
plt.pie(values, labels=label, autopct = custom_autopct, startangle=90, counterclock=False, colors=color, wedgeprops=wedgeprops, pctdistance=0.7) 
plt.show()
```

## DataFrame활용

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
# 학교별 비율 원 그래프 그려보기
wedgeprops = {'width':0.6, 'edgecolor':'k', 'linewidth':2}   

grp = df.groupby('학교')

values = [ grp.size()['북산고'], grp.size()['능남고'] ]   # [5,3]입력한 셈 
label = [ '북산고', '능남고' ]

def custom_autopct(pct):
    return ('%0.1f%%' % pct) if pct >= 10 else '' 


plt.pie(values, labels=label, autopct = custom_autopct, startangle=90, counterclock=False, colors=color, wedgeprops=wedgeprops, pctdistance=0.7) 
plt.title('소속 학교')
plt.show()
```
