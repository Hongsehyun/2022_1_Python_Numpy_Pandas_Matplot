# 16. 여러그래프

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 사용 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # Windows에서 한글 사용
matplotlib.rcParams['font.size'] = 15                  # 한글 글자 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
 # fig, axs = 2개의 데이터를 받아오겠다는 의미
 # subplot(2, 2) = 2 x 2 로 총 4개의 그래프를 찍겠다는 의미
fig, axs = plt.subplots(2, 2, figsize=(7.5, 5))
fig.suptitle('여러 그래프 넣기')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10))   # 2 x 2 로 총 4개의 그래프를 찍겠다는 의미
fig.suptitle('여러 그래프 넣기')

# 첫번째 그래프 넣기
axs[0,0].bar(df['이름'], df['국어'], label='국어점수')   # 데이터 설정
axs[0,0].set_title('첫 번째 그래프')                     # 제목 설정
axs[0,0].legend()                                        # 범례 표시
axs[0,0].set(xlabel='이름', ylabel='점수')               # x, y축 label 입력
axs[0,0].set_facecolor('lightyellow')                    # 전면 색 설정
axs[0,0].grid(linestyle='--', linewidth=0.5)             # 그리드 표시

# 두 번째 그래프 넣기
axs[0,1].plot(df['이름'], df['수학'], label='수학점수')   # 데이터 설정
axs[0,1].plot(df['이름'], df['영어'], label='영어점수')   # 데이터 설정
axs[0,1].legend()                                        # 범례 표시

# 세 번째 그래프 넣기
axs[1,0].barh(df['이름'], df['키'])   # 데이터 설정

# 네 번째 그래프 넣기
axs[1,1].plot(df['이름'], df['사회'], label='사회점수', color='green', alpha=1)   # 데이터 설정
```
