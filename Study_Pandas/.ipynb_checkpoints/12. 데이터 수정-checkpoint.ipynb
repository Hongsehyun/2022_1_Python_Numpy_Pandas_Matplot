# 12. 데이터 수정

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

## Column 수정

```python
# [방법1] replace함수를 통해 '북산고->상북고', '능남고->무슨고'로 이름을 바꿔줌!
df['학교'].replace( {'북산고':'상북고', '능남고':'무슨고'}, inplace=True )
df
```

```python
# [방법2] 한 열의 내용을 모두 소문자로 바꿔줌!
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
# [방법3] 기존에 존재하는 데이터에 내용 추가하는 방법!
df['학교'] = df['학교'] + '등학교'
df
```

***

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail'   # 결과 Column을 추가하고, 전체 데이터는 Fail로 초기화

df.loc[df['총합']>400, '결과'] = 'Pass'   # 총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

***

# Column 삭제

```python
df.drop(columns=['총합'])  # 총합 Column삭제
```

```python
df.drop(columns=['국어','영어','수학'])  # 총합 Column삭제
```

***

# Row 삭제

```python
# 인덱스 번호에 따라 Row 삭제하기 
df.drop(index='4번')
```

```python
# 조건에 맞는 Row만 삭제하기
filt = df['수학'] < 80
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index)
```

***

## Row 추가하기

```python
df.loc['9번'] = ['홍세현', '동산고등학교', 170, 90, 90, 90, 90, 90, 'Matlab', 450, 'Pass']
df
```

***

## Cell 수정하기

```python
df.loc['4번', 'SW특기'] = 'Python'   # 4번 학생의 데이터를 1개만 변경 (SW특기 내용에 Python입력)
df
```

```python
df.loc['5번', ['학교','SW특기']] = ['동산고등학교','C']   # 5번 학생의 데이터를 동시에 2개를 변경
df
```

***

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
# 결과 Column을 맨 앞으로 댕겨오기
df = df[[cols[-1]] + cols[0:-1]]
df
```
