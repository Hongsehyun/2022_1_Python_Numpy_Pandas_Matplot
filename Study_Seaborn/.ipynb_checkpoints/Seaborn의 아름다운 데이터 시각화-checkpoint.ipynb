```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

sns.set(style='darkgrid')   # seaborn의 서식 미리 지정

#%matplotlib inline          # 자동 사이즈 조정
```

```python
# 데이터셋 불러오기
tips = sns.load_dataset('tips')
tips
```

```python
# Seaborn으로 그래프 Plotting하기
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# Yes/No로만 값을 가지는 colormap 추가 [hue]
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
# 다양한 숫자형 데이터 값으로만 값을 가지는 colormap 추가 [size]
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
# style & hue로 2개의 데이터 동시에 출력하여 보이기
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
# 색상 - Seaborn Palette 색상 Library 구글에 검색하면 종류 나옴!
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='rocket')
```

```python
# Size - 마커의 size바꾸기
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='rocket')
```
